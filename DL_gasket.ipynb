```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from matplotlib import pyplot as plt
```

```python
df_flange = pd.read_csv('data/gasket.csv')
```

```python
df_flange_data = df_flange.query("T == 'training'")
```

```python
df_flange_train = df_flange.query("T == 'training'").drop(columns=['T','No','30','40','50','60','70','80','90'])
df_flange_test = df_flange.query("T == 'test'").drop(columns=['T','No','30','40','50','60','70','80','90'])
```

```python
df_flange_30 = df_flange_data['30']
df_flange_40 = df_flange_data['40']
df_flange_50 = df_flange_data['50']
df_flange_60 = df_flange_data['60']
df_flange_70 = df_flange_data['70']
df_flange_80 = df_flange_data['80']
df_flange_90 = df_flange_data['90']
```

# 소요시점 별 Model 세팅

```python
model_30 = Sequential()
model_30.add(Dense(20, input_dim = 13, kernel_initializer='normal', activation='relu'))
model_30.add(Dense(1, kernel_initializer = 'normal'))
model_30.compile(loss='mse', optimizer='adam')

model_40 = Sequential()
model_40.add(Dense(20, input_dim = 13, kernel_initializer='normal', activation='relu'))
model_40.add(Dense(1, kernel_initializer = 'normal'))
model_40.compile(loss='mse', optimizer='adam')

model_50 = Sequential()
model_50.add(Dense(20, input_dim = 13, kernel_initializer='normal', activation='relu'))
model_50.add(Dense(1, kernel_initializer = 'normal'))
model_50.compile(loss='mse', optimizer='adam')

model_60 = Sequential()
model_60.add(Dense(20, input_dim = 13, kernel_initializer='normal', activation='relu'))
model_60.add(Dense(1, kernel_initializer = 'normal'))
model_60.compile(loss='mse', optimizer='adam')

model_70 = Sequential()
model_70.add(Dense(20, input_dim = 13, kernel_initializer='normal', activation='relu'))
model_70.add(Dense(1, kernel_initializer = 'normal'))
model_70.compile(loss='mse', optimizer='adam')

model_80 = Sequential()
model_80.add(Dense(20, input_dim = 13, kernel_initializer='normal', activation='relu'))
model_80.add(Dense(1, kernel_initializer = 'normal'))
model_80.compile(loss='mse', optimizer='adam')

model_90 = Sequential()
model_90.add(Dense(20, input_dim = 13, kernel_initializer='normal', activation='relu'))
model_90.add(Dense(1, kernel_initializer = 'normal'))
model_90.compile(loss='mse', optimizer='adam')
```

# 트레이닝

```python
hist_30 = model_30.fit(df_flange_train,df_flange_30,epochs = 500, batch_size=5,verbose=0)
hist_40 = model_40.fit(df_flange_train,df_flange_40,epochs = 500, batch_size=5,verbose=0)
hist_50 = model_50.fit(df_flange_train,df_flange_50,epochs = 500, batch_size=5,verbose=0)
hist_60 = model_60.fit(df_flange_train,df_flange_60,epochs = 500, batch_size=5,verbose=0)
hist_70 = model_70.fit(df_flange_train,df_flange_70,epochs = 500, batch_size=5,verbose=0)
hist_80 = model_80.fit(df_flange_train,df_flange_80,epochs = 500, batch_size=5,verbose=0)
hist_90 = model_90.fit(df_flange_train,df_flange_90,epochs = 500, batch_size=5,verbose=0)
```

# loss 그래프

```python
plt.figure(figsize=(14,10))
plt.plot(hist_30.history['loss'], label='30')
plt.plot(hist_40.history['loss'], label='40')
plt.plot(hist_50.history['loss'], label='50')
plt.plot(hist_60.history['loss'], label='60')
plt.plot(hist_70.history['loss'], label='70')
plt.plot(hist_80.history['loss'], label='80')
plt.plot(hist_90.history['loss'], label='90')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(loc='best')
plt.show()
```

# loss evaluation

```python
loss_30 = model_30.evaluate(df_flange_train, df_flange_30, batch_size=5)
print('loss_30: '+ str(loss_30))
loss_40 = model_40.evaluate(df_flange_train, df_flange_40, batch_size=5)
print('loss_40: '+ str(loss_40))
loss_50 = model_50.evaluate(df_flange_train, df_flange_50, batch_size=5)
print('loss_50: '+ str(loss_50))
loss_60 = model_60.evaluate(df_flange_train, df_flange_60, batch_size=5)
print('loss_60: '+ str(loss_60))
loss_70 = model_70.evaluate(df_flange_train, df_flange_70, batch_size=5)
print('loss_70: '+ str(loss_70))
loss_80 = model_80.evaluate(df_flange_train, df_flange_80, batch_size=5)
print('loss_80: '+ str(loss_80))
loss_90 = model_90.evaluate(df_flange_train, df_flange_90, batch_size=5)
print('loss_90: '+ str(loss_90))
```
