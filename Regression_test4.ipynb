```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
test = pd.read_csv('data/source/test.csv')
x_test = pd.DataFrame(test, columns=['x'])
y_test = pd.DataFrame(test, columns=['y'])
x_train = pd.DataFrame(test, columns=['c'])
```

```python
linear_regressor = LinearRegression()
linear_model = linear_regressor.fit(x_test,y_test)
linear_predict = linear_model.predict(x_train)
```

```python
poly_regressor2 = Pipeline([('poly', PolynomialFeatures(degree=2)),('linear', LinearRegression(fit_intercept=False))])
poly_model2 = poly_regressor2.fit(x_test,y_test)
poly_predict2 = poly_model2.predict(x_train)
```

```python
plt.figure(figsize=(14,10))
plt.scatter(test.x, test.y, label = 'data')
plt.plot(x_train, linear_predict, label = 'linear')
plt.plot(x_train, poly_predict2, label = '2nd degree')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc = 'best')
plt.title("regression test")
plt.show()
```
