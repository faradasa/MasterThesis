```python
from apyori import apriori
import pyfpgrowth
import numpy as np
import pandas as pd
import csv
from matplotlib import pyplot as plt
```

```python
f = open('data/data3.txt','r')
```

```python
transaction = []
```

```python
for i, line in enumerate(f):
    print(line.strip('\n'))
    text = line.strip('\n')
    array = text.split(',')
    array = list(map(int,array))
    transaction.append(array)
```

```python
transaction
```

# apriori

```python
result = list(apriori(transaction))
print(result)
```

# FPgrowth

```python
patterns = pyfpgrowth.find_frequent_patterns(transaction,2)
rules = pyfpgrowth.generate_association_rules(patterns,0.02)
```
