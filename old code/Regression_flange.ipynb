# PKG import, data input

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
flange30 = pd.read_csv('data/flange_30.csv')
flange40 = pd.read_csv('data/flange_40.csv')
flange50 = pd.read_csv('data/flange_50.csv')
flange60 = pd.read_csv('data/flange_60.csv')
flange70 = pd.read_csv('data/flange_70.csv')
flange80 = pd.read_csv('data/flange_80.csv')
flange90 = pd.read_csv('data/flange_90.csv')
```

```python
x_flange30 = flange30.drop(columns = ['Flange_usage'])
x_flange40 = flange40.drop(columns = ['Flange_usage'])
x_flange50 = flange50.drop(columns = ['Flange_usage'])
x_flange60 = flange60.drop(columns = ['Flange_usage'])
x_flange70 = flange70.drop(columns = ['Flange_usage'])
x_flange80 = flange80.drop(columns = ['Flange_usage'])
x_flange90 = flange90.drop(columns = ['Flange_usage'])
```

```python
y_flange30 = pd.DataFrame(flange30, columns = ['Flange_usage'])
y_flange40 = pd.DataFrame(flange40, columns = ['Flange_usage'])
y_flange50 = pd.DataFrame(flange50, columns = ['Flange_usage'])
y_flange60 = pd.DataFrame(flange60, columns = ['Flange_usage'])
y_flange70 = pd.DataFrame(flange70, columns = ['Flange_usage'])
y_flange80 = pd.DataFrame(flange80, columns = ['Flange_usage'])
y_flange90 = pd.DataFrame(flange90, columns = ['Flange_usage'])
```

```python
flange_test = pd.read_csv('data/flange_test.csv')
```

# Linear regression

```python
linear_regressor_30 = LinearRegression()
linear_regressor_40 = LinearRegression()
linear_regressor_50 = LinearRegression()
linear_regressor_60 = LinearRegression()
linear_regressor_70 = LinearRegression()
linear_regressor_80 = LinearRegression()
linear_regressor_90 = LinearRegression()
```

```python
linear_model_flange30 = linear_regressor_30.fit(x_flange30,y_flange30)
linear_model_flange40 = linear_regressor_40.fit(x_flange40,y_flange40)
linear_model_flange50 = linear_regressor_50.fit(x_flange50,y_flange50)
linear_model_flange60 = linear_regressor_60.fit(x_flange60,y_flange60)
linear_model_flange70 = linear_regressor_70.fit(x_flange70,y_flange70)
linear_model_flange80 = linear_regressor_80.fit(x_flange80,y_flange80)
linear_model_flange90 = linear_regressor_90.fit(x_flange90,y_flange90)
```

```python
flange30_prediction_linear = linear_model_flange30.predict(flange_test)
flange40_prediction_linear = linear_model_flange40.predict(flange_test)
flange50_prediction_linear = linear_model_flange50.predict(flange_test)
flange60_prediction_linear = linear_model_flange60.predict(flange_test)
flange70_prediction_linear = linear_model_flange70.predict(flange_test)
flange80_prediction_linear = linear_model_flange80.predict(flange_test)
flange90_prediction_linear = linear_model_flange90.predict(flange_test)
```

```python
print(flange30_prediction_linear)
print(flange40_prediction_linear)
print(flange50_prediction_linear)
print(flange60_prediction_linear)
print(flange70_prediction_linear)
print(flange80_prediction_linear)
print(flange90_prediction_linear)
```

# 2차 poly regression
* degree 부분을 3으로 수정하면 3차 에 대한 regression

```python
poly30 = Pipeline([('poly', PolynomialFeatures(degree=2)),('linear', LinearRegression(fit_intercept=False))])
poly40 = Pipeline([('poly', PolynomialFeatures(degree=2)),('linear', LinearRegression(fit_intercept=False))])
poly50 = Pipeline([('poly', PolynomialFeatures(degree=2)),('linear', LinearRegression(fit_intercept=False))])
poly60 = Pipeline([('poly', PolynomialFeatures(degree=2)),('linear', LinearRegression(fit_intercept=False))])
poly70 = Pipeline([('poly', PolynomialFeatures(degree=2)),('linear', LinearRegression(fit_intercept=False))])
poly80 = Pipeline([('poly', PolynomialFeatures(degree=2)),('linear', LinearRegression(fit_intercept=False))])
poly90 = Pipeline([('poly', PolynomialFeatures(degree=2)),('linear', LinearRegression(fit_intercept=False))])
```

```python
poly_model_flange30 = poly30.fit(x_flange30,y_flange30)
poly_model_flange40 = poly40.fit(x_flange40,y_flange40)
poly_model_flange50 = poly50.fit(x_flange50,y_flange50)
poly_model_flange60 = poly60.fit(x_flange60,y_flange60)
poly_model_flange70 = poly70.fit(x_flange70,y_flange70)
poly_model_flange80 = poly80.fit(x_flange80,y_flange80)
poly_model_flange90 = poly90.fit(x_flange90,y_flange90)  
```

```python
flange30_prediction_poly = poly_model_flange30.predict(flange_test)
flange40_prediction_poly = poly_model_flange40.predict(flange_test)
flange50_prediction_poly = poly_model_flange50.predict(flange_test)
flange60_prediction_poly = poly_model_flange60.predict(flange_test)
flange70_prediction_poly = poly_model_flange70.predict(flange_test)
flange80_prediction_poly = poly_model_flange80.predict(flange_test)
flange90_prediction_poly = poly_model_flange90.predict(flange_test)
```

```python
print(flange30_prediction_poly)
print(flange40_prediction_poly)
print(flange50_prediction_poly)
print(flange60_prediction_poly)
print(flange70_prediction_poly)
print(flange80_prediction_poly)
print(flange90_prediction_poly)  
```

# 시계열로 비교

```python
flange_time = pd.read_csv('data/flange_time.csv')
time_test = pd.read_csv('data/time_test.csv')
```

```python
x_flange_time = pd.DataFrame(flange_time, columns = ['time'])
y_flange_time = pd.DataFrame(flange_time, columns = ['flange_usage'])
```

```python
linear_regressor_time = LinearRegression()
linear_model_time = linear_regressor_time.fit(x_flange_time,y_flange_time)
time_linear_predict = linear_model_time.predict(time_test)
```

```python
poly_time = Pipeline([('poly', PolynomialFeatures(degree=2)),('linear', LinearRegression(fit_intercept=False))])
poly_model_time = poly_time.fit(x_flange_time,y_flange_time)
time_poly_predict = poly_model_time.predict(time_test)
```

```python
poly_time3 = Pipeline([('poly', PolynomialFeatures(degree=3)),('linear', LinearRegression(fit_intercept=False))])
poly_model_time3 = poly_time3.fit(x_flange_time,y_flange_time)
time_poly_predict3 = poly_model_time3.predict(time_test)
```

```python
poly_time4 = Pipeline([('poly', PolynomialFeatures(degree=4)),('linear', LinearRegression(fit_intercept=False))])
poly_model_time4 = poly_time4.fit(x_flange_time,y_flange_time)
time_poly_predict4 = poly_model_time4.predict(time_test)
```

```python
poly_time5 = Pipeline([('poly', PolynomialFeatures(degree=5)),('linear', LinearRegression(fit_intercept=False))])
poly_model_time5 = poly_time5.fit(x_flange_time,y_flange_time)
time_poly_predict5 = poly_model_time5.predict(time_test)
```

```python
plt.figure(figsize = (14,10))
plt.scatter(flange_time.time, flange_time.flange_usage, label = "actual")
plt.plot(time_test.time,time_linear_predict, label = "Linear")
plt.plot(time_test.time, time_poly_predict, label = "2 degree")
plt.plot(time_test.time, time_poly_predict3, label = "3 degree")
plt.plot(time_test.time, time_poly_predict4, label = "4 degree")
plt.plot(time_test.time, time_poly_predict5, label = "5 degree")
plt.xlabel("time")
plt.ylabel("flange_usage")
plt.legend(loc="best")
plt.title("Regressions")
plt.show()
```
